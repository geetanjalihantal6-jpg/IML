```python
# Question 2
def check(n):
  return(n>0)-(n<0)
def answer(n):
  return{-1:'Negative',0:'Zero',1:'Positive'}[check(n)]
n= int(input('Enter n: '))
print(n,'is', answer(n))
```

```python
#Question 3: Factorial
n= int(input('Enter n: '))
fact=1
if n<0:
  print('Enter positive number: ')
else:
  for i in range(1,n+1):
    fact=fact*i
  print(f'Factorial of {n} is, {fact}')
```

```python
# Question 4: prime Number

n= int(input('Enter n: '))
c= 0
if n==0:
  print('0 is neither positive nor negative')
elif n<0:
  print('Enter positive number')
else:
  for i in range(1,n+1):
    if n%i==0:
      c+=1
  if c==2:
    print(n,'is prime')
  else:
    print(n,'is not prime')
```
