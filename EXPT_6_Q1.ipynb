```python
import numpy as np
from sklearn.cluster import KMeans

# Define the data points
points = {
    "A": (2, 3),
    "B": (3, 4),
    "C": (6, 6),
    "D": (7, 7)
}

labels = list(points.keys())
data = np.array(list(points.values()))

# Print Input Data Points
print("Input Data Points:")
for i in range(len(labels)):
    print(f"{labels[i]} : {data[i]}")

# Apply K-Means clustering
kmeans = KMeans(n_clusters=2, random_state=0, n_init=10)
kmeans.fit(data)

cluster_labels = kmeans.labels_
centroids = kmeans.cluster_centers_

# Assign points to clusters based on which centroid is C1 and C2 in the desired output
# Desired output: C1 has centroid [2.5 3.5] and C2 has centroid [6.5 6.5]
# Based on `centroids = [[6.5 6.5], [2.5 3.5]]`
# This means cluster_label 1 (for centroid [2.5 3.5]) should be C1
# And cluster_label 0 (for centroid [6.5 6.5]) should be C2
C1 = []
C2 = []

for i in range(len(cluster_labels)):
    if cluster_labels[i] == 1: # cluster_label 1 corresponds to centroid [2.5 3.5]
        C1.append(labels[i])
    else: # cluster_label 0 corresponds to centroid [6.5 6.5]
        C2.append(labels[i])

# Print Centroid Values (matching C1 to centroids[1] and C2 to centroids[0])
print("\nCentroid Values:")
print(f"Centroid 1 (C1): {centroids[1]}") # Centroid for cluster_label 1
print(f"Centroid 2 (C2): {centroids[0]}") # Centroid for cluster_label 0

# Print Cluster Assignments
print("\nCluster Assignments:")
print(f"C1 (Centroid 1): {C1}")
print(f"C2 (Centroid 2): {C2}")

# Print points under each cluster
print("\nPoints under C1:")
for p in C1:
    print(f"{p} -> {points[p]}")

print("\nPoints under C2:")
for p in C2:
    print(f"{p} -> {points[p]}")
```
