```python
import numpy as np
from sklearn.linear_model import LinearRegression

# Given data points
x = np.array([1, 2, 3]).reshape(-1, 1) # Reshape for scikit-learn
y = np.array([2, 3, 5])

# Create a Linear Regression model
model = LinearRegression()

# Fit the model to the data
model.fit(x, y)

# Get the coefficients and intercept
intercept = model.intercept_
coefficient = model.coef_[0]

print(f"Intercept (b0): {intercept:.2f}")
print(f"Coefficient (b1): {coefficient:.2f}")
print(f"The linear regression equation is: y = {coefficient:.2f}x + {intercept:.2f}")

# Predict y values based on the model
y_pred = model.predict(x)
print(f"Predicted y values: {y_pred}")
```

```python
import matplotlib.pyplot as plt

plt.scatter(x, y, color='blue', label='Actual data points')
plt.plot(x, y_pred, color='red', label='Regression line')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Regression for x=[1,2,3], y=[2,3,5]')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from sklearn.linear_model import Ridge

# Create a Ridge Regression model with a small alpha (lambda)
ridge_model = Ridge(alpha=0.1) # alpha is the lambda parameter in scikit-learn

# Fit the model to the data
ridge_model.fit(x, y)

# Get the coefficients and intercept for Ridge Regression
ridge_intercept = ridge_model.intercept_
ridge_coefficient = ridge_model.coef_[0]

print(f"Ridge Intercept (b0): {ridge_intercept:.2f}")
print(f"Ridge Coefficient (b1): {ridge_coefficient:.2f}")
print(f"The Ridge Regression equation is: y = {ridge_coefficient:.2f}x + {ridge_intercept:.2f}")

# Predict y values based on the Ridge model
y_pred_ridge = ridge_model.predict(x)

plt.scatter(x, y, color='blue', label='Actual data points')
plt.plot(x, y_pred, color='red', linestyle='--', label='OLS Regression line')
plt.plot(x, y_pred_ridge, color='green', label='Ridge Regression line (alpha=0.1)')
plt.xlabel('x')
plt.ylabel('y')
plt.title('OLS vs Ridge Regression')
plt.legend()
plt.grid(True)
plt.show()
```
