```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.naive_bayes import CategoricalNB
data = {
    'Weather': ['sunny', 'sunny', 'rainy', 'rainy'],
    'PlayGame': ['yes', 'yes', 'no', 'no']
}
df = pd.DataFrame(data)

display(df)

le_weather = LabelEncoder()
le_playgame = LabelEncoder()

df['Weather_encoded'] = le_weather.fit_transform(df['Weather'])
df['PlayGame_encoded'] = le_playgame.fit_transform(df['PlayGame'])

X = df[['Weather_encoded']]
y = df['PlayGame_encoded']

print("Encoded Weather Mapping:", list(le_weather.classes_))
print("Encoded PlayGame Mapping:", list(le_playgame.classes_))
display(df)

model = CategoricalNB()
model.fit(X, y)

print("Model trained successfully.")
weather_to_predict = 'sunny'

# Encode the input weather
encoded_weather_to_predict = le_weather.transform([weather_to_predict])

# Make prediction
prediction_encoded = model.predict([encoded_weather_to_predict])

# Decode the prediction back to original label
prediction_label = le_playgame.inverse_transform(prediction_encoded)

print(f"When the weather is '{weather_to_predict}', the game will be played: {prediction_label[0]}")
```
