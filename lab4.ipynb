```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([2, 3, 4])
y = np.array([1, 2, 3])

lam = 1

X = np.column_stack((np.ones(len(x)), x))

I = np.eye(X.shape[1])


beta = np.linalg.inv(X.T @ X + lam * I) @ X.T @ y
b0, b1 = beta

print("b0 =", b0)
print("b1 =", b1)

x_plot = np.linspace(1.5, 4.5, 100)
y_plot = b0 + b1 * x_plot

plt.scatter(x, y)
plt.plot(x_plot, y_plot)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Ridge Regression (Linear)")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt


x1 = np.array([1, 2, 3])
x2 = np.array([2, 3, 4])
y = np.array([1, 2, 3])


X = np.column_stack((np.ones(len(x1)), x1, x2))

lam = 1

I = np.eye(X.shape[1])

beta = np.linalg.inv(X.T @ X + lam * I) @ X.T @ y

b0, b1, b2 = beta

print("b0 =", b0)
print("b1 =", b1)
print("b2 =", b2)

y_pred = X @ beta

plt.scatter(y, y_pred)
plt.plot(y, y, linestyle='--')
plt.xlabel("Actual y")
plt.ylabel("Predicted y")
plt.title("Ridge Regression: Actual vs Predicted")
plt.show()
```
