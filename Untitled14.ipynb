```python
import numpy as np

# Given values
alpha = np.array([1, 1, 1])
support_vectors = np.array([
    [0, -1, 1],
    [0, 2, -1],
    [-1, 0, 2]
])
y_labels = np.array([1, -1, -1])
input_feature_vector = np.array([0.2, 0.8, 0.4])

print(f"Input feature vector (x): {input_feature_vector}\n")

# 1. Calculate the weight vector 'w'
# The weight vector w is given by the sum of (alpha_i * y_i * x_i) for all support vectors.
w = np.sum(alpha[:, np.newaxis] * y_labels[:, np.newaxis] * support_vectors, axis=0)
print(f"Calculated weight vector (w): {w}")

# 2. Calculate the bias 'b'
# For a hard margin SVM, each support vector x_k satisfies y_k * (w . x_k + b) = 1.
# This implies b = y_k - (w . x_k).
# Due to the hypothetical nature of the problem, the values for 'b' derived from each
# support vector are inconsistent. To address this, we will average these 'b' values.
b_values = []
for i in range(len(support_vectors)):
    x_k = support_vectors[i]
    y_k = y_labels[i]
    dot_product_w_xk = np.dot(w, x_k)
    b_k = y_k - dot_product_w_xk
    b_values.append(b_k)

print(f"Individual 'b' values derived from each support vector: {b_values}")
b = np.mean(b_values)
print(f"Averaged bias (b): {b}\n")

# 3. Compute the decision function output for the input feature vector
# f(x) = w . x + b
decision_function_output = np.dot(w, input_feature_vector) + b
print(f"Decision function output (f(x)): {decision_function_output}")

# 4. Determine the predicted class label
# The predicted class label is the sign of f(x).
predicted_class_label = np.sign(decision_function_output)

print(f"Predicted class label (y): {int(predicted_class_label)}")
```
