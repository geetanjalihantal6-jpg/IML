```python
import numpy as np

data_points = np.array([2, 4, 10, 12, 3, 20, 30, 11, 25])
k = 2

# Initial centroids
c1 = 2
c2 = 20
centroids = np.array([c1, c2])

print(f"Data points: {data_points}")
print(f"Initial centroids: {centroids}")
```

```python
# Function to assign points to the nearest centroid
def assign_to_clusters(data, centroids):
    clusters = [[] for _ in range(len(centroids))]
    for point in data:
        distances = [np.abs(point - centroid) for centroid in centroids]
        closest_centroid_idx = np.argmin(distances)
        clusters[closest_centroid_idx].append(point)
    return clusters

# Function to update centroids
def update_centroids(clusters):
    new_centroids = []
    for cluster in clusters:
        if len(cluster) > 0:
            new_centroids.append(np.mean(cluster))
        else:
            new_centroids.append(np.nan) # Handle empty clusters if necessary
    return np.array(new_centroids)

# First iteration
print("\n--- Iteration 1 ---")
clusters_iter1 = assign_to_clusters(data_points, centroids)
print("Cluster assignments after 1st iteration:")
for i, cluster in enumerate(clusters_iter1):
    print(f"  Cluster {i+1} (centroid {centroids[i]}): {cluster}")

new_centroids_iter1 = update_centroids(clusters_iter1)
print(f"New centroids after 1st iteration: {new_centroids_iter1}")

centroids = new_centroids_iter1 # Update centroids for the next iteration
```

```python
# Second iteration (and potential convergence)
print("\n--- Iteration 2 ---")
clusters_iter2 = assign_to_clusters(data_points, centroids)
print("Cluster assignments after 2nd iteration:")
for i, cluster in enumerate(clusters_iter2):
    print(f"  Cluster {i+1} (centroid {centroids[i]}): {cluster}")

new_centroids_iter2 = update_centroids(clusters_iter2)
print(f"New centroids after 2nd iteration: {new_centroids_iter2}")

if np.array_equal(centroids, new_centroids_iter2):
    print("Centroids did not change. K-Means converged.")
else:
    print("Centroids changed. Continuing if not converged.")
    centroids = new_centroids_iter2

# Final Clusters (assuming convergence after 2 iterations or showing the state after 2)
print("\n--- Final Clusters ---")
for i, cluster in enumerate(clusters_iter2):
    print(f"  Cluster {i+1}: {cluster} (Centroid: {new_centroids_iter2[i]:.2f})")
```

## Refactor K-Means function

### Subtask:
Encapsulate the K-Means clustering logic (assignment and update steps) into a single, structured function that takes data points, k, and initial centroids as input, and returns the final clusters and centroids.


**Reasoning**:
To encapsulate the K-Means clustering logic into a single function as requested, I will define a new Python function `kmeans_clustering` which will take data, k, and initial centroids as input, and will iteratively assign points to clusters and update centroids until convergence or max_iterations is reached. The function will then return the final clusters and centroids.



```python
def kmeans_clustering(data, k, initial_centroids, max_iterations=100):
    centroids = np.array(initial_centroids)

    for i in range(max_iterations):
        # Step 1: Assign points to the nearest centroid
        clusters = assign_to_clusters(data, centroids)

        # Step 2: Update centroids
        new_centroids = update_centroids(clusters)

        # Check for convergence
        if np.array_equal(centroids, new_centroids):
            print(f"K-Means converged after {i + 1} iterations.")
            break

        centroids = new_centroids
    else:
        print(f"K-Means reached max iterations ({max_iterations}) without converging.")

    return clusters, centroids

print("Defined the kmeans_clustering function.")
```

**Reasoning**:
The `kmeans_clustering` function has been defined. Now, I will call this function with the `data_points`, `k`, and `initial_centroids` to execute the K-Means clustering and store the results.



```python
final_clusters, final_centroids = kmeans_clustering(data_points, k, np.array([c1, c2]))

print("\n--- K-Means Clustering Results (using kmeans_clustering function) ---")
print(f"Final Centroids: {final_centroids}")
print("Final Cluster Assignments:")
for i, cluster in enumerate(final_clusters):
    print(f"  Cluster {i+1} (centroid {final_centroids[i]:.2f}): {cluster}")
```
