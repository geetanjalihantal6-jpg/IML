```python
import numpy as np
from sklearn.linear_model import Ridge

# Define the input data
X = np.array([2, 3, 4]).reshape(-1, 1)  # Reshape X to a 2D array for sklearn
y = np.array([1, 2, 3])

# Define the regularization parameter (alpha)
alpha = 1.0

# Create a Ridge Regression model instance
ridge_model = Ridge(alpha=alpha)

# Fit the model to the data
ridge_model.fit(X, y)

# Get the coefficients and intercept
coefficient = ridge_model.coef_[0]
intercept = ridge_model.intercept_

print(f"Calculated Coefficient: {coefficient:.4f}")
print(f"Calculated Intercept: {intercept:.4f}")
```
