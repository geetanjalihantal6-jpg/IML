```python
#APPLY AGGLOMERATIVE APPROACH TO CREATE THE DENDROGRAM

import numpy as np
from scipy.cluster.hierarchy import dendrogram, linkage
import matplotlib.pyplot as plt

# Define the data points
data = np.array([
    [1, 1],  # Point A
    [2, 1],  # Point B
    [4, 3],  # Point C
    [5, 4]   # Point D
])

# Define labels for the points
labels = ['A', 'B', 'C', 'D']

# Perform agglomerative clustering using 'single' linkage method
# 'single' linkage method uses the minimum distance between observations of the two sets
linked = linkage(data, method='single')

# Plot the dendrogram
plt.figure(figsize=(8, 5))
dendrogram(linked,
           orientation='top',
           labels=labels,
           distance_sort='aescending',
           show_leaf_counts=True)
plt.title('Agglomerative Clustering Dendrogram (Single Linkage)')
plt.xlabel('Data Point')
plt.ylabel('Distance')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
