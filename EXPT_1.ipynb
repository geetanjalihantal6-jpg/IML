```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, r2_score

# Given datapoints
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 5, 4, 5])

# Perform linear regression using numpy.polyfit (least squares method)
# np.polyfit(x, y, 1) returns coefficients [slope, intercept] for a 1st degree polynomial
coefficients = np.polyfit(x, y, 1)

slope = coefficients[0]
intercept = coefficients[1]

print(f"Regression line equation: y = {slope:.2f}x + {intercept:.2f}")

# Generate predicted y values based on the regression line
y_predicted = slope * x + intercept

# Calculate error metrics
mse = mean_squared_error(y, y_predicted)
r2 = r2_score(y, y_predicted)

print(f"Mean Squared Error (MSE): {mse:.2f}")
print(f"R-squared (R2) Score: {r2:.2f}")

# Plotting the data and the regression line
plt.figure(figsize=(8, 6))
plt.scatter(x, y, color='blue', label='Actual Data Points')
plt.plot(x, y_predicted, color='red', label='Regression Line')
plt.title('Linear Regression: Actual vs. Predicted')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid(True)
plt.show()
```
